```python
import tensorflow as tf
from jax.example_libraries.optimizers import optimizer
from tensorflow import keras
import numpy as np
```

# Define the model

```python
model = keras.Sequential([keras.layers.Dense(1,input_dim=2, activation = 'sigmoid')])
optimizer = keras.optimizers.SGD(learning_rate=0.4)
model.compile(optimizer = optimizer, loss = 'binary_crossentropy')
```

# Define the data

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]])

Y = np.array([0,1,1,1]) # OR function
```

# Train the model

```python
history = model.fit(X,Y, epochs = 200 ) #verbose = 0
```

# print the final loss

```python
final_loss  = history.history['loss'][-1]
print(F"Final loss: {final_loss}")
```

# print the predictions

```python
for x ,y in X:
    prediction = model.predict(np.array([[x,y]]))
    print(f"Input: {x},{y}, Predicted Output: {prediction[0][0]}")
```
